```python
import pandas as pd

df = pd.read_excel('bls2024_occupation_national.xlsx')
df
```

```python
# Explore data structure
print("Dataset shape:", df.shape)
print("\nColumn names:")
print(df.columns.tolist())
print("\nData types:")
print(df.dtypes)
print("\nFirst few rows:")
print(df.head())
```

```python
# Let's examine columns more carefully
print(f"Total columns: {len(df.columns)}")
print("\nColumns containing 'occup', 'job', 'work', or 'employ':")
job_related_cols = [col for col in df.columns if any(keyword in col.lower() for keyword in ['occup', 'job', 'work', 'employ'])]
for col in job_related_cols:
    print(f"- {col}")

print("\nColumns containing 'income', 'wage', 'salary', 'earn', or 'pay':")
income_related_cols = [col for col in df.columns if any(keyword in col.lower() for keyword in ['income', 'wage', 'salary', 'earn', 'pay'])]
for col in income_related_cols:
    print(f"- {col}")
```

```python
# Let's look at all column names to understand the data better
print("All columns in the dataset:")
for i, col in enumerate(df.columns, 1):
    print(f"{i:2d}. {col}")

print(f"\nDataset has {df.shape[0]:,} rows and {df.shape[1]} columns")
```

```python
# Examine key columns for job analysis
print("Key columns for our analysis:")
print("- OCC_CODE: Occupation code")
print("- OCC_TITLE: Occupation title (job profile)")
print("- TOT_EMP: Total employment")
print("- H_MEAN: Mean hourly wage")
print("- A_MEAN: Mean annual wage")
print("- H_MEDIAN: Median hourly wage")
print("- A_MEDIAN: Median annual wage")

print("\nSample occupation titles:")
print(df['OCC_TITLE'].head(10).tolist())

print("\nChecking for missing values in key columns:")
key_cols = ['OCC_TITLE', 'TOT_EMP', 'H_MEAN', 'A_MEAN', 'H_MEDIAN', 'A_MEDIAN']
for col in key_cols:
    missing = df[col].isna().sum()
    print(f"{col}: {missing} missing values")
```

```python
# Clean the data for analysis
# Filter out broad occupational categories and focus on specific jobs
# Remove rows where employment is very low (less than 100) to focus on substantial occupations
df_clean = df.copy()

# Remove broad categories (usually have "Occupations" in the title)
df_specific = df_clean[~df_clean['OCC_TITLE'].str.contains('Occupations', na=False)]

# Remove entries with very low employment (less than 100 people)
df_specific = df_specific[df_specific['TOT_EMP'] >= 100]

# Convert wage columns to numeric, replacing any non-numeric values with NaN
wage_cols = ['H_MEAN', 'A_MEAN', 'H_MEDIAN', 'A_MEDIAN']
for col in wage_cols:
    df_specific[col] = pd.to_numeric(df_specific[col], errors='coerce')

# Remove rows where wage data is missing
df_specific = df_specific.dropna(subset=['A_MEAN', 'A_MEDIAN'])

print(f"After filtering: {len(df_specific)} specific occupations with substantial employment and wage data")
print(f"Original dataset: {len(df)} rows")
print(f"Removed: {len(df) - len(df_specific)} rows")
```

```python
# Find the highest paying jobs
print("=== TOP 20 HIGHEST PAYING JOBS BY ANNUAL MEAN SALARY ===")
highest_paying = df_specific.nlargest(20, 'A_MEAN')[['OCC_TITLE', 'A_MEAN', 'A_MEDIAN', 'TOT_EMP']]

for i, (idx, row) in enumerate(highest_paying.iterrows(), 1):
    title = row['OCC_TITLE']
    mean_salary = row['A_MEAN']
    median_salary = row['A_MEDIAN']
    employment = row['TOT_EMP']
    
    print(f"{i:2d}. {title}")
    print(f"    Mean Annual Salary: ${mean_salary:,.0f}")
    print(f"    Median Annual Salary: ${median_salary:,.0f}")
    print(f"    Total Employment: {employment:,.0f}")
    print()
```

```python
# Find the lowest paying jobs
print("=== TOP 20 LOWEST PAYING JOBS BY ANNUAL MEAN SALARY ===")
lowest_paying = df_specific.nsmallest(20, 'A_MEAN')[['OCC_TITLE', 'A_MEAN', 'A_MEDIAN', 'TOT_EMP']]

for i, (idx, row) in enumerate(lowest_paying.iterrows(), 1):
    title = row['OCC_TITLE']
    mean_salary = row['A_MEAN']
    median_salary = row['A_MEDIAN']
    employment = row['TOT_EMP']
    
    print(f"{i:2d}. {title}")
    print(f"    Mean Annual Salary: ${mean_salary:,.0f}")
    print(f"    Median Annual Salary: ${median_salary:,.0f}")
    print(f"    Total Employment: {employment:,.0f}")
    print()
```

```python
# Summary statistics and comparisons
print("=== SALARY DISTRIBUTION SUMMARY ===")
print(f"Total occupations analyzed: {len(df_specific):,}")
print()

# Summary statistics
print("Annual Mean Salary Statistics:")
print(f"Minimum: ${df_specific['A_MEAN'].min():,.0f}")
print(f"Maximum: ${df_specific['A_MEAN'].max():,.0f}")
print(f"Average: ${df_specific['A_MEAN'].mean():,.0f}")
print(f"Median: ${df_specific['A_MEAN'].median():,.0f}")
print()

# Salary ranges
high_paying = df_specific[df_specific['A_MEAN'] >= 100000]
low_paying = df_specific[df_specific['A_MEAN'] <= 30000]

print(f"High-paying jobs ($100k+ annually): {len(high_paying):,} ({len(high_paying)/len(df_specific)*100:.1f}%)")
print(f"Low-paying jobs ($30k or less annually): {len(low_paying):,} ({len(low_paying)/len(df_specific)*100:.1f}%)")

# Employment in high vs low paying jobs
total_emp_high = high_paying['TOT_EMP'].sum()
total_emp_low = low_paying['TOT_EMP'].sum()
total_emp_all = df_specific['TOT_EMP'].sum()

print()
print("Employment Distribution:")
print(f"People in high-paying jobs: {total_emp_high:,.0f} ({total_emp_high/total_emp_all*100:.1f}%)")
print(f"People in low-paying jobs: {total_emp_low:,.0f} ({total_emp_low/total_emp_all*100:.1f}%)")
```

```python
# Let's create visualizations
import matplotlib.pyplot as plt
import numpy as np

# Create figure with subplots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))

# 1. Salary distribution histogram
ax1.hist(df_specific['A_MEAN'], bins=50, alpha=0.7, color='skyblue', edgecolor='black')
ax1.set_xlabel('Annual Mean Salary ($)')
ax1.set_ylabel('Number of Occupations')
ax1.set_title('Distribution of Annual Mean Salaries')
ax1.grid(True, alpha=0.3)

# 2. Top 15 highest paying jobs
top_15_high = df_specific.nlargest(15, 'A_MEAN')
y_pos = np.arange(len(top_15_high))
ax2.barh(y_pos, top_15_high['A_MEAN'], color='green', alpha=0.7)
ax2.set_yticks(y_pos)
ax2.set_yticklabels([title[:40] + '...' if len(title) > 40 else title for title in top_15_high['OCC_TITLE']], fontsize=8)
ax2.set_xlabel('Annual Mean Salary ($)')
ax2.set_title('Top 15 Highest Paying Jobs')
ax2.grid(True, alpha=0.3)

# 3. Bottom 15 lowest paying jobs
bottom_15_low = df_specific.nsmallest(15, 'A_MEAN')
y_pos = np.arange(len(bottom_15_low))
ax3.barh(y_pos, bottom_15_low['A_MEAN'], color='red', alpha=0.7)
ax3.set_yticks(y_pos)
ax3.set_yticklabels([title[:40] + '...' if len(title) > 40 else title for title in bottom_15_low['OCC_TITLE']], fontsize=8)
ax3.set_xlabel('Annual Mean Salary ($)')
ax3.set_title('Bottom 15 Lowest Paying Jobs')
ax3.grid(True, alpha=0.3)

# 4. Salary vs Employment scatter plot
ax4.scatter(df_specific['TOT_EMP'], df_specific['A_MEAN'], alpha=0.6, s=30)
ax4.set_xlabel('Total Employment')
ax4.set_ylabel('Annual Mean Salary ($)')
ax4.set_title('Salary vs Employment')
ax4.set_xscale('log')  # Log scale for better visibility
ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
